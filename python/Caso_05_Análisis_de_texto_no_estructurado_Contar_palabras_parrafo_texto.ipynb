### Contar palabras

Se utiliza la función count() para crear una lista de la cantidad de ocasiones que hay de cada palabra



```python
frecuencia_palabras = []
for p in lista_completa:
    frecuencia_palabras.append(lista_completa.count(p))


print("Frecuenciasde palabras: ", "\n", lista_completa, "\n")
print("Cantidad de elementos de la lista completa:", "\n", len(lista_completa), "\n")
print("Frecuenciasde palabras: ", "\n", frecuencia_palabras, "\n")
print("Cantidad de elementos de la frecuencia de palabras: ", "\n", len(frecuencia_palabras), "\n")


```

```python
import pandas as pd
import matplotlib.pyplot as plt

texto = """
Python es un lenguaje de programación de alto nivel, interpretado y de propósito general.
Fue creado por Guido van Rossum y lanzado por primera vez en 1991.
Python se ha vuelto increíblemente popular en los últimos años debido a su sintaxis simple y legible,
lo que lo hace ideal para principiantes y expertos por igual.
"""

# Convertir el texto en una lista de palabras
palabras = texto.split()

# Contar la frecuencia de cada palabra
frecuencia_palabras = {}
for palabra in palabras:
    palabra = palabra.lower()  # Convertir la palabra a minúsculas para evitar contar palabras con mayúsculas y minúsculas como diferentes
    if palabra in frecuencia_palabras:
        frecuencia_palabras[palabra] += 1
    else:
        frecuencia_palabras[palabra] = 1

# Imprimir la frecuencia de las palabras
for palabra, frecuencia in frecuencia_palabras.items():
    print(f'{palabra}: {frecuencia}')

# Crear el DataFrame
df = pd.DataFrame({'palabra': list(frecuencia_palabras.keys()), 'frecuencia': list(frecuencia_palabras.values())})

# Mostrar el DataFrame
print("\nDataFrame:")
print(df)
print()

# Graficar el diagrama de barras
plt.figure(figsize=(10, 6))
plt.bar(df['palabra'], df['frecuencia'], color='skyblue')
plt.xlabel('Palabra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de palabras')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import re
from collections import Counter

import requests
import chardet

# Función para cargar archivo texto
def obtener_contenido(url):
    response = requests.get(url)
    if response.status_code == 200:
        # Obtener el contenido como bytes
        contenido_bytes = response.content

        # Detectar la codificación
        codificacion = chardet.detect(contenido_bytes)['encoding']

        # Decodificar el contenido con la codificación detectada
        contenido = contenido_bytes.decode(codificacion)

        return contenido
    else:
        return None

# Función para procesar el texto y obtener la frecuencia de palabras
def frecuencia_palabras(texto):
    # Eliminar signos de puntuación
    texto = re.sub(r'[^\w\s]', '', texto)
    # Convertir a minúsculas
    texto = texto.lower()
    # Eliminar artículos, conectores y palabras comunes
    palabras_comunes = {'el', 'la', 'los', 'las', 'un', 'una', 'unos', 'unas', 'y', 'o', 'pero', 'a', 'de', 'en', 'con', 'para', 'por', 'si'}
    palabras = [palabra for palabra in texto.split() if palabra not in palabras_comunes]
    # Calcular la frecuencia de cada palabra
    frecuencia = Counter(palabras)
    return frecuencia

# URL del archivo de texto
url = 'https://raw.githubusercontent.com/rpizarrog/analitica_de_datos/main/datos/Anteproyecto%20HERRAMIENTa%20DE%20REGRESI%C3%93N%20multivariable.txt'

# Obtener el contenido del archivo de texto
contenido = obtener_contenido(url)

if contenido:
    # Obtener la frecuencia de palabras
    frecuencia = frecuencia_palabras(contenido)

    # Crear DataFrame con las palabras y su frecuencia
    df = pd.DataFrame(list(frecuencia.items()), columns=['Palabra', 'Frecuencia'])

    # Ordenar el DataFrame por frecuencia y seleccionar las 10 palabras más frecuentes
    df_top10 = df.sort_values(by='Frecuencia', ascending=False).head(10)

    # Crear diagrama de barras con las 10 palabras más frecuentes
    plt.figure(figsize=(10, 6))
    plt.bar(df_top10['Palabra'], df_top10['Frecuencia'], color='skyblue')
    plt.xlabel('Palabra')
    plt.ylabel('Frecuencia')
    plt.title('Top 10 Palabras Más Frecuentes')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Crear nube de palabras con todas las palabras
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frecuencia)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Nube de Palabras')
    plt.show()
else:
    print('No se pudo obtener el contenido del archivo.')
```
